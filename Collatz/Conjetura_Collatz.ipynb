# CONJETURA DE COLLATZ

```python
n = int(input("Ingresa unnùmoer entero"))
count = 0
if n == 1:
    print (n)
while n > 1:
    if n%2 == 0:
        n = int(n/2)
        count +=1
        print(n)
    elif n%2 != 0:
        n = int(3*n + 1)
        count += 1
        print (n)
print ("Pasos = ", count)
```
