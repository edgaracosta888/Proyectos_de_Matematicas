# 1. FUNCIÓN QUE DETERMINA SI UN NÚMERO n ES O NO PRIMO

```python
# ======================================================================================
# Determina si un número entero n es primo.
# Devuelve un valor booleano:
#     True  -> si n es primo
#     False -> si n no es primo
# ======================================================================================
def Prime(n: int | None = None) -> bool:
# ======================================================================================
# Validación sobre n. Si n no es un número entero, se muestra un mensaje de error y
# retorna False.
# ======================================================================================
    try:
        n = int(n)
    except (ValueError, TypeError):
        print("El argumento debe ser un número.")
        return False
# ======================================================================================    
# Los números menores que 2 no se consideran primos.
# ======================================================================================
    if n < 2:
        return False
# ======================================================================================
# Basta revisar divisores hasta la raíz cuadrada de n. Retorna False si se encuentra
# algún divisor de n en el rango desde 2 hasta la raíz cuadrada de n.
# ======================================================================================
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
# ======================================================================================
# Retorna True si n es primo, es decir, si no se encontró ningún divisor de n en el rango
# desde 2 hasta la raíz cuadrada de n.
# ======================================================================================
    return True
```

# 2. FUNCIÓN QUE ENCUENTRA TODOS LOS NÚMEROS PRIMOS ENTRE UN RANGO (m, n).

```python
# ======================================================================================
# Genera una lista de números primos entre dos límites m y n.
# Devuelve una lista con los números primos encontrados.
# ======================================================================================

def List_Prime_numbers(m: int | None = None, n: int | None = None) -> list:
# ======================================================================================
# Validación sobre m y n. Si m o n no son números enteros, se muestra un mensaje de
# error y se retorna una lista vacía.
# ======================================================================================
   try:
      m = int(m)
      n = int(n)
   except (ValueError, TypeError):
      print("Los argumentos deben ser números enteros.")
      return []
# ======================================================================================
# Validación de los límites m y n.
# Si m es mayor que n, se muestra un mensaje de error.
# ======================================================================================
   if int(m) > int(n):
      print("El límite superior debe ser mayor o igual al límite inferior.")
      return []
# ======================================================================================
# Validación de los límites m y n. 
# Si n es menor que 2, se devuelve una lista vacía,ya que no hay números primos menores
# que 2. Si m es menor que 2, se ajusta a 2, ya que el primer número primo es 2.
# ======================================================================================
   elif int(n) < 2:
      return []
# ======================================================================================
# Validación de los límites m y n.
# Si m es menor que 2, se ajusta a 2, ya que el primer número primo es 2.
# ======================================================================================
   elif int(m) < 2:
      m = 2

   m = int(m)
   n = int(n)

# ======================================================================================
# Genera una lista de números primos entre dos límites m y n.
# ======================================================================================
   P = []
# ======================================================================================
# Validación sobre m.
# Si m es igual a 2, se agrega a la lista de primos.
# ======================================================================================
   if m == 2:
      P.append(m)
# ======================================================================================
# Validación sobre m.
# Si m es par mayor que 2, se deja m sin cambios pues se ajustará en el bucle.
# ======================================================================================
   elif m > 2 and m%2 == 0:
      m = m
# ======================================================================================
# Validación sobre m.
# Si m es impar mayor que 2, se ajusta a m-1 pues en el bucle tomará de nuevo su valor.
# ======================================================================================
   else:
      m = m - 1
# ======================================================================================
# Verifica si en el rango entre m y n hay números primos, y si los hay, los agrega a la 
# lista P.
# ======================================================================================      
   for i in range(m + 1, n + 1, 2):
      if Prime(i):
         P.append(i)

   return P
```

# 3. FUNCIÓN QUE ENCUENTRA EL NÚMERO PRIMO INMEDIATAMENTE POSTERIOR A UN NÚMERO DADO n.

```python
# ======================================================================================
# Función que devuelve el número primo siguiente a un número entero n.
# ======================================================================================
def Next_Prime(n: int | None = None) -> int | None:
# ======================================================================================
# Validación sobre n. Si n no es un número entero, se muestra un mensaje de error y se
# retorna None.
# ======================================================================================
    try:
        n = int(n)
    except (ValueError, TypeError):
        print("El argumento debe ser un número.")
        return None
# ======================================================================================
# Validación sobre n.
# Si n es menor que 2, devuelve 2.
# ======================================================================================
    if n < 2:
        return 2
# ======================================================================================
# Validación sobre n.
# Si n es par, se ajusta a n+1 y se busca el siguiente número primo sumando 2 en cada
# iteración.
# ======================================================================================
    if n % 2 == 0:
        n += 1
        s = 0
        while Prime (n + s) == False:
            s += 2
        next = n + s
# ======================================================================================
# Validación sobre n.
# Si n es impar, se busca el siguiente número primo sumando 2 en cada iteración.
# ======================================================================================
    else:
        s = 2
        while Prime (n + s) == False:
            s += 2
        next = n + s

    return (next)
```

# 4. FUNCIÓN QUE ENCUENTRA EL NÚMERO PRIMO INMEDIATAMENTE ANTERIOR A UN NÚMERO DADO n.

```python
# ======================================================================================
# Función que devuelve el número primo anterior a un número entero n.
# ======================================================================================
def Previous_Prime(n: int | None = None) -> int | None:
# ======================================================================================
# Validación sobre n. Si n no es un número entero, se muestra un mensaje de error y se
# retorna None.
# ======================================================================================
    try:
        n = int(n)
    except (ValueError, TypeError):
        print("El argumento debe ser un número.")
        return None
# ======================================================================================
# Validación sobre n.
# Si n es menor o igual a 2, imprime un mensaje indicando que el número primo más 
# pequeño es 2, y retorna None.
# ======================================================================================
    if n <= 2:
        print("El número primo más pequeño es 2.")
        return None
# ======================================================================================
# Validación sobre n. 
# Si n es par, se ajusta a n-1 y se busca el anterior número primo restando 2 en cada
# iteración.
# ======================================================================================
    if n % 2 == 0:
        n -= 1
        s = 0
        while Prime (n - s) == False:
            s += 2
        next = n - s
# ======================================================================================
# Validación sobre n.
# Si n es impar, se busca el siguiente número primo sumando 2 en cada iteración.
# ======================================================================================
    else:
        s = 2
        while Prime (n - s) == False:
            s += 2
        next = n - s

    return (next)
```

# 5. FUNCIÓN QUE ENCUENTRA LOS NÚMEROS PRIMOS INMEDIATAMENTE ANTERIOR Y POSTERIOR PARA UN NUMERO DADO n.

```python
# ======================================================================================
# Función que devuelve el número primo anterior y el siguiente a un número entero n.
# ======================================================================================
def Previous_Next_Prime(n: int | None = None) -> tuple:

# ======================================================================================
# Validación sobre n. Si n no es un número entero, se muestra un mensaje de error y se
# retorna None.
# ======================================================================================
    try:
        n = int(n)
    except (ValueError, TypeError):
        print("El argumento debe ser un número.")
        return None
# ======================================================================================
# Se obtiene el número primo anterior y el siguiente a n utilizando las funciones 
# Previous_Prime y Next_Prime, respectivamente.
# ======================================================================================
    previous = Previous_Prime(n)
    next = Next_Prime(n)

    return (previous, next)
```

# 6. FUNCIÓN QUE ENCUENTRA LOS PRIMEROS n NÚMEROS PRIMOS.

```python
# ======================================================================================
# Función que devuelve una lista con los primeros n números primos.
# ======================================================================================
def First_Prime_Numbers(n):
# ======================================================================================
# Validación sombre n. Si n no es un número entero, se muestra un mensaje de error y se
# retorna None.
# ======================================================================================
    try:
        n = int(n)
    except (ValueError, TypeError):
        print("El argumento debe ser un número entero.")
        return None

# ======================================================================================
# Validación sobre n.
# Si n es menor que 1, devuelve una lista vacía.
# ======================================================================================
    if n < 1:
        return []

# ======================================================================================
# Validación sobre n.
# Si n es 1, devuelve una lista con el primer número primo, que es 2.
# ======================================================================================
    elif n == 1:
        return [2]

# ======================================================================================
# Generación de los primeros n números primos.
# Se usa la función Next_Primo para obtener el siguiente número primo a partir del 
# último número primo encontrado (comenzando con 2), hasta obtener n números primos en total.
# ======================================================================================
    else:
        p = [2]
        contador = len(p)
        while contador < n:
            p.append(Next_Prime(p[-1]))
            contador = contador + 1
        return p
```
