```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def Primo(n):
    # Los números menores o iguales a 1 no son primos
    if n <= 1:
        return False
    # El 2 y el 3 son primos
    if n <= 3:
        return True
    # Eliminar los múltiplos de 2 y 3
    if n % 2 == 0 or n % 3 == 0:
        return False
    # Comprobar divisibilidad hasta la raíz cuadrada de n
    limite = int(math.sqrt(n)) + 1
    for i in range(5, limite, 6):  # Verifica 5, 7, 11, 13, 17...
        if n % i == 0 or n % (i + 2) == 0:
            return False
    return True
```

```python
def Lista_Primos(m = 0, n = 0):
   
   P = []
   if m != 0 and n == 0:
      for i in range(m+1):
         if i == 2 or i%2 != 0 and Primo(i) == True:
            P.append(i)
      return P
   else:
      for i in range(m,n+1):
         if i == 2 or i%2 != 0 and Primo(i) == True:
            P.append(i)
      return P
```

```python
def Previous_Next_Primo(n):

    if n <= 2:
        previous = np.nan
    else:
        s = 1
        while Primo(n - s) == False:
            s = s + 1
        previous = n - s


    s = 1
    while Primo (n + s) == False:
        s = s + 1
    next = n + s

    return (previous, next)
```

```python
def Next_Primo(n):

    s = 1
    while Primo (n + s) == False:
        s = s + 1
    next = n + s

    return (next)
```

```python
def Pirmeros_Primos(n):
    if n == 1:
        return [2]
    p = [2]
    contador = len(p)
    while contador < n:
        p.append(Next_Primo(p[-1]))
        contador = contador + 1
    return p
```
